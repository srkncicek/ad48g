```r
# I import required packages. I will use vehicles dataset from fuelconomy package.

library(data.table)
install.packages("fueleconomy")

library(fueleconomy)

dt <- as.data.table(vehicles)
```

```r
# I check and see my dt is data.table
class(dt)
```

```r
# I look at the first rows of the dt
head(dt)
```

```r
# I do some i operations by filtering the dt 

dt[make %in% c("Audi", "BMW", "Volkswagen", "Tesla") & year > 2000 & hwy > 15]
```

```r
# I did some j operations using slicing method. make feature to year.
dt[make %in% c("Audi", "BMW", "Volkswagen", "Tesla") & year > 2000 & hwy > 15, make:year ]
```

```r
# I did j operations again but now I assign new column which is calculate and add %20 percentage tp cyl (cylinders) feature. I know it is not so meaningfull
dt[make %in% c("Audi", "BMW", "Volkswagen", "Tesla") & year > 2000 & hwy > 15,
   new_cyl := cyl * 1.20]
```

```r
# We can see new column here
dt[make %in% c("Audi", "BMW", "Volkswagen", "Tesla") & year > 2000 & hwy > 15]
```

```r
# In this step I did by operations and calculate max min and av value of year for each make.
dt[, .(max_at_o = max(year, na.rm = T), min_at_o = min(year, na.rm = T), av_at_o = mean(year, na.rm = T)), by = make]
```

```r
# I show make, class and max, min, and avg year for corresponding make and class.
dt_at <- dt[, .(max_at_oc = max(year, na.rm = T), min_at_oc = min(year, na.rm = T), av_at_oc = mean(year, na.rm = T)),
        by = c("make", "class")]
```

```r
# we can see table here
dt_at
```

```r
# I use cast function in this cell for wide form.In this table we can see average  year for associated combination.
dt_at_wide <- dcast(dt_at, make ~ class, value.var = "av_at_oc")
```

```r
dt_at_wide
```

```r
# I transformed previous table into long form again using melt function.
melt(dt_at_wide, id.vars = "make", variable.name = "class_new", value.name = "av_time_new")
```
